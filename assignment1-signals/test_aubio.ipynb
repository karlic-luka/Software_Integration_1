```python
import aubio
import numpy as np
import librosa
import os
```

```python
sample_rate=44100
x=np.zeros(44100)
for i in range(44100):
    x[i]=np.sin(2. * np.pi * i * 225. / sample_rate)

# create pitch object
p = aubio.pitch("yin", samplerate = sample_rate)
# other examples:
# = aubio.pitch("yinfft", 4096, 512, 44100)
# = aubio.pitch("yin", 2048, 512, 44100)
# = aubio.pitch("mcomb", 4096, 512, 44100)
# = aubio.pitch("schmitt", samplerate = 44100, hop_size = 512, buf_size = 2048)

# pad end of input vector with zeros
pad_length = p.hop_size - x.shape[0] % p.hop_size
x_padded = np.pad(x, (0, pad_length), 'constant', constant_values=0)
# to reshape it in blocks of hop_size
x_padded = x_padded.reshape(-1, p.hop_size)

# input array should be of type aubio.float_type (defaults to float32)
x_padded = x_padded.astype(aubio.float_type)

for frame, i in zip(x_padded, range(len(x_padded))):
    time_str = "%.3f" % (i * p.hop_size/float(sample_rate))
    pitch_candidate = p(frame)[0]
    print (time_str, "%.3f" % pitch_candidate)
```

```python
# Constants for the audio file
BUFFER_SIZE = 1024
CHANNELS = 1
RATE = 44100

# Path to the .wav file
path = 'audio_inputs'
file = 'piano-a_A_major.wav'
filename = os.path.join(path, file)

# Create an Aubio pitch detection object
methods = ["default", "yinfft", "yin", "mcomb", "schmitt"]
idx = 3
pDetection = aubio.pitch(methods[idx], BUFFER_SIZE,
                         BUFFER_SIZE, RATE) # bilo je BUFFER_SIZE//2 kao treci parametar
pDetection.set_unit("Hz")
# pDetection.set_silence(-40)
pDetection.set_tolerance(0.9)

# Use aubio.source to read the .wav file
source = aubio.source(filename, RATE, BUFFER_SIZE)

while True:
    try:
        # Read data from the .wav file
        samples, read = source()

        # Estimate the pitch
        pitch = pDetection(samples)[0]

        # Print the estimated pitch
        print(pitch)

        # Break the loop if end of file reached
        if read < BUFFER_SIZE:
            break
    except KeyboardInterrupt:
        break
```

```python
import pyaudio
import aubio

# Constants for the audio stream
BUFFER_SIZE = 1024
CHANNELS = 1
RATE = 44100

# Create a PyAudio object
p = pyaudio.PyAudio()

# Create an audio stream
stream = p.open(format=pyaudio.paFloat32,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=BUFFER_SIZE)

# Create an Aubio pitch detection object
methods = ["default", "yinfft", "yin", "mcomb", "schmitt"]
idx = 2
pDetection = aubio.pitch(methods[idx], BUFFER_SIZE,
                         BUFFER_SIZE, RATE)
pDetection.set_unit("Hz")
pDetection.set_silence(-40)

while True:
    try:
        # Read data from the audio stream
        data = stream.read(BUFFER_SIZE)
        samples = np.fromstring(data, dtype=aubio.float_type)

        # Estimate the pitch
        pitch = pDetection(samples)[0]

        print(pitch)
    except KeyboardInterrupt:
        break

# Close the stream
stream.stop_stream()
stream.close()
p.terminate()
```
